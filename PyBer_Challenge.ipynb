### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

urban_cities_df_1 = city_data_df[city_data_df["type"] == "Urban"]
suburban_cities_df_1 = city_data_df[city_data_df["type"] == "Suburban"]
rural_cities_df_1 = city_data_df[city_data_df["type"] == "Rural"]
```

```python
#  1. Get the total rides for each city type

urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"].sum()
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"].sum()
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"].sum()
```

```python
# 2. Get the total drivers for each city type
urban_driver_count = urban_cities_df_1.groupby(["city"]).sum()["driver_count"].sum()


suburban_driver_count = suburban_cities_df_1.groupby(["city"]).sum()["driver_count"].sum()


rural_driver_count = rural_cities_df_1.groupby(["city"]).sum()["driver_count"].sum()
urban_driver_count
```

```python
#  3. Get the total amount of fares for each city type
urban_fare_count = urban_cities_df.groupby(["city"]).sum()["fare"].sum()

suburban_fare_count = suburban_cities_df.groupby(["city"]).sum()["fare"].sum()

rural_fare_count = rural_cities_df.groupby(["city"]).sum()["fare"].sum()
urban_fare_count
```

```python
#  4. Get the average fare per ride for each city type. 
average_urban_fare_per_ride = urban_fare_count / urban_ride_count

average_suburban_fare_per_ride = suburban_fare_count / suburban_ride_count

average_rural_fare_per_ride = rural_fare_count / rural_ride_count
average_rural_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_urban_fare_per_driver = urban_fare_count /urban_driver_count

average_suburban_fare_per_driver =suburban_fare_count /suburban_driver_count

average_rural_fare_per_driver = rural_fare_count /rural_driver_count
```

```python
city_type = pyber_data_df["type"].unique()
city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": [rural_ride_count, suburban_ride_count, urban_ride_count],
    "Total Drivers": [rural_driver_count, suburban_driver_count, urban_driver_count],
    "Total Fares": [rural_fare_count, suburban_fare_count, urban_fare_count],
    "Average Fare per Ride": [average_rural_fare_per_ride, average_suburban_fare_per_ride, average_urban_fare_per_ride],
    "Average Fare per Driver": [average_rural_fare_per_driver ,average_suburban_fare_per_driver, average_rural_fare_per_driver]})

pyber_summary_df.index = ["Rural","Suburban", "Urban"]    
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df.info()
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_pyber_data_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
new_pyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
new_pyber_data_df = new_pyber_data_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_pyber_data_df_piv = new_pyber_data_df.pivot(index="date", columns="type", values="fare")
new_pyber_data_df_piv
```

```python
new_pyber_data_df_piv.info()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range = new_pyber_data_df_piv.loc[(new_pyber_data_df_piv["date"] >= 2019-01-01) & (new_pyber_data_df_piv["date"] <= 2019-04-28)] 
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
#df.index = pd.to_datetime(df.index)
new_pyber_data_df_piv.index = pd.to_datetime(new_pyber_data_df_piv.index)
new_pyber_data_df_piv
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_pyber_data_df_piv.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_by_week = new_pyber_data_df_piv.resample("W").sum()
sum_of_fares_by_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')



sum_of_fares_by_week.plot(kind ='line', figsize=(20,6), title='Total Fare by City Type', ylabel='Fare($USD)', xlabel='')

plt.savefig("analysis/PyBer_fare_summary.png")
```
